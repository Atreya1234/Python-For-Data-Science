# Question 5: Plot Tesla Stock Graph

```python
Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph. The structure to call the make_graph function
is make_graph(tesla_data, tesla_revenue, 'Tesla')
```

```python
make_graph(tesla_data, tesla_revenue, 'Tesla')
```

```python
# Assuming tesla_data and tesla_revenue are already defined in your context
# Example datasets (you should replace these with actual data)

# Example of tesla_data
tesla_data = {
    'Date': ['2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01', '2021-05-01'],
    'Close': [700, 800, 850, 780, 900]
}

# Example of tesla_revenue
tesla_revenue = {
    'Date': ['2021-01-01', '2021-02-01', '2021-03-01', '2021-04-01', '2021-05-01'],
    'Revenue': [10000, 12000, 15000, 13000, 16000]
}

# Define the make_graph function (assuming this is part of your codebase)
import matplotlib.pyplot as plt

def make_graph(stock_data, revenue_data, company):
    fig, ax1 = plt.subplots()

    color = 'tab:red'
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Price', color=color)
    ax1.plot(stock_data['Date'], stock_data['Close'], color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  
    color = 'tab:blue'
    ax2.set_ylabel('Revenue', color=color)
    ax2.plot(revenue_data['Date'], revenue_data['Revenue'], color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    plt.title(f'{company} Stock Price and Revenue')
    fig.tight_layout()  
    plt.show()

# Now, call the make_graph function with the Tesla data
make_graph(tesla_data, tesla_revenue, 'Tesla')

```

# Question 6: Plot GameStop Stock Graph

```python
Use the make_graph function to graph the GameStop Stock Data, also provide a title for the graph. The structure to call the make_graph
function is make_graph(gme_data, gme_revenue, 'GameStop').
```

```python
make_graph(gme_data, gme_revenue, 'GameStop')
```

```python
# Example data for gme_data and gme_revenue
gme_data = {
    'Date': ['2021-01-01', '2021-02-01', '2021-03-01'],
    'Stock Price': [17.25, 325.00, 120.00]
}

gme_revenue = {
    'Date': ['2020-10-31', '2021-01-31', '2021-04-30'],
    'Revenue': [1297.0, 2112.0, 1227.0]
}

# Define the make_graph function (assuming it's provided by the system you're working with)
def make_graph(stock_data, revenue_data, title):
    import matplotlib.pyplot as plt
    import pandas as pd

    # Convert dictionaries to DataFrames
    stock_df = pd.DataFrame(stock_data)
    revenue_df = pd.DataFrame(revenue_data)

    # Convert Date columns to datetime
    stock_df['Date'] = pd.to_datetime(stock_df['Date'])
    revenue_df['Date'] = pd.to_datetime(revenue_df['Date'])

    # Create a figure and a set of subplots
    fig, ax1 = plt.subplots()

    # Plot stock data
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Stock Price', color='tab:blue')
    ax1.plot(stock_df['Date'], stock_df['Stock Price'], color='tab:blue')
    ax1.tick_params(axis='y', labelcolor='tab:blue')

    # Create a second y-axis to plot revenue data
    ax2 = ax1.twinx()
    ax2.set_ylabel('Revenue ($M)', color='tab:red')
    ax2.plot(revenue_df['Date'], revenue_df['Revenue'], color='tab:red')
    ax2.tick_params(axis='y', labelcolor='tab:red')

    # Add title and show plot
    plt.title(title)
    fig.tight_layout()
    plt.show()

# Call the make_graph function with the title 'GameStop Stock and Revenue Data'
make_graph(gme_data, gme_revenue, 'GameStop Stock and Revenue Data')
```
